Pour mes observations, je vais pivoter la table en gardant uniquement les valeurs de 2010 pour rendre ça plus clair et simple à analyser

````python
# Pivoter la table en utilisant pivot
pivoted_data = df.pivot(index='Country Name', columns='Indicator Name', values='2010')
pivoted_data = pivoted_data.reset_index()
pivoted_data.shape
````

Nous avons désormais un dataframe plus simple à analyser, cependant il y a maintenant 3666 colonnes liés à chaque indicateurs, il faut donc réduire cela. Pour se faire, je vais filtrer les données de la même façon que pour les années et regarder parmis les indiateurs avec le plus de valeurs pour sélectionner mon dataframe final.

````python
non_null_counts = pivoted_data.drop('Country Name', axis=1).count()
sorted_columns = non_null_counts.sort_values(ascending=False)
top_columns = sorted_columns.head(50)
plt.figure(figsize=(16, 12))
top_columns.plot(kind='barh', color='steelblue')
plt.xlabel('Nombre de Valeurs Non Nulles')
plt.ylabel('Indicateurs')
plt.title('Indicateurs avec le Plus de Valeurs Non Nulles (année 2010)')
plt.gca().invert_yaxis()
plt.show()
````

Ici nous avons une idée des indicateurs les plus représentés et pouvant potentiellement correspondre à nos besoins, je vais utiliser cette liste pour en sélectionner un plus petit échantillon parmis ceux qui me semblent les plus pertinents pour la problématique.
Voici la liste d'indicateurs que je souhaite conserver :
- Gross outbound enrolment ratio, all regions, both sexes (%) = correspond au taux d'engagement des étudiants quittant leur pays pour aller étudier à l'étranger, peut être pertinent pour indiquer un intérêt pour les programmes d'éducation en ligne internationaux.
- GDP per capita (current US \$\) = Les pays avec un PIB par habitant élevé peuvent avoir une population disposée à investir dans l'éducation, y compris l'éducation en ligne.
- Internet users (per 100 people) = Un accès Internet répandu peut faciliter l'adoption de l'éducation en ligne
- GNI per capita, PPP (current international \$\) = Le Revenu national brut (RNB) par habitant ajusté en parité de pouvoir d'achat (PPP). Cela peut être pertinent pour évaluer la capacité financière de la population à accéder à des services d'éducation en ligne.

````python
selected_columns = [
    "Gross outbound enrolment ratio, all regions, both sexes (%)",
    "GDP per capita (current US$)",
    "Internet users (per 100 people)",
    "GNI per capita, PPP (current international $)",
    "Population of the official age for upper secondary education, both sexes (number)",
    "Population, total"
]

df_filtre = pivoted_data[['Country Name'] + selected_columns]

print(df_filtre.head())
````

## Normalisation des données pour obtenir un score avec les meilleurs pays

Maintenant que nous avons pu observer les top pays pour chacuns de nos indicateurs, nous pouvons essayer de normaliser tous ça et tenter de calculer un score global pour voir quels pays semblent remplir au mieux tous les critères que nous avons défini

````python
# attribution d'un poids pour chacuns des indicateurs.
weights = {
    "Gross outbound enrolment ratio, all regions, both sexes (%)": 0.1,
    "GDP per capita (current US$)": 0.2,
    "Internet users (per 100 people)": 0.2,
    "GNI per capita, PPP (current international $)": 0.2,
    "Population of the official age for upper secondary education, both sexes (number)": 0.2,
    "Population, total": 0.1
}

indicators = list(weights.keys())
df_indicators = df_filtre[indicators]

# normalisation des données pour le score
df_normalized = (df_indicators - df_indicators.min()) / (df_indicators.max() - df_indicators.min()) * 10
df_normalized.columns = ["Normalized " + col for col in df_normalized.columns]

df_score = pd.concat([df_filtre, df_normalized], axis=1)

# Calcul du score en multipliant chaque indicateur par son poids, puis en sommant les résultats
df_score["Score"] = (df_normalized * list(weights.values())).sum(axis=1)
df_score = df_score.sort_values(by="Score", ascending=False)
top_10_pays = df_score.head(10)

plt.figure(figsize=(10, 6))
plt.barh(top_10_pays["Country Name"], top_10_pays["Score"], color='steelblue')
plt.xlabel('Score')
plt.ylabel('Pays')
plt.title('Top 10 des Pays en Fonction du Score')
plt.gca().invert_yaxis()
plt.show()
````
